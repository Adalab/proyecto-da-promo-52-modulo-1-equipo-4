# ğŸ¾ Juego de Preguntas sobre Animales (VersiÃ³n WHILE)

Este juego tipo test pone a prueba tus conocimientos sobre animales.

- Cada pregunta tiene 4 opciones (A, B, C, D)
- Solo una respuesta es correcta
- El juego termina cuando:
  - Acumulas **5 aciertos** (ganas)
  - Acumulas **3 errores** (pierdes)
  - O se acaban las preguntas disponibles


```python
import random

# Lista de preguntas
preguntas = [
    {
        "pregunta": "Â¿CuÃ¡l de estos animales es un mamÃ­fero?",
        "opciones": {"A": "PingÃ¼ino", "B": "DelfÃ­n", "C": "TiburÃ³n", "D": "Cocodrilo"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "Â¿QuÃ© animal tiene la lengua mÃ¡s larga en proporciÃ³n a su cuerpo?",
        "opciones": {"A": "Jirafa", "B": "CamaleÃ³n", "C": "Hormiga", "D": "MurciÃ©lago"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "Â¿CuÃ¡l de estos animales pone huevos?",
        "opciones": {"A": "Gato", "B": "Perro", "C": "Gallina", "D": "Caballo"},
        "respuesta_correcta": "C"
    },
    {
        "pregunta": "Â¿QuÃ© animal es conocido por su memoria?",
        "opciones": {"A": "Elefante", "B": "Cebra", "C": "RatÃ³n", "D": "Tortuga"},
        "respuesta_correcta": "A"
    },
    {
        "pregunta": "Â¿CuÃ¡l es el Ãºnico mamÃ­fero capaz de volar?",
        "opciones": {"A": "Pato", "B": "MurciÃ©lago", "C": "Avestruz", "D": "Gaviota"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "Â¿QuÃ© animal puede cambiar de color para camuflarse?",
        "opciones": {"A": "TiburÃ³n", "B": "Pulpo", "C": "CamaleÃ³n", "D": "Gato"},
        "respuesta_correcta": "C"
    },
    {
        "pregunta": "Â¿CuÃ¡l de estos animales es un reptil?",
        "opciones": {"A": "Rana", "B": "Serpiente", "C": "PingÃ¼ino", "D": "Ballena"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "Â¿QuÃ© animal tiene una sola pierna cuando camina?",
        "opciones": {"A": "Canguro", "B": "Flamenco", "C": "Pulpo", "D": "Caracol"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "Â¿QuÃ© animal duerme de pie?",
        "opciones": {"A": "Gato", "B": "Caballo", "C": "Loro", "D": "Oso"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "Â¿QuÃ© animal tiene el corazÃ³n mÃ¡s grande en proporciÃ³n a su cuerpo?",
        "opciones": {"A": "Ballena azul", "B": "Elefante", "C": "RatÃ³n", "D": "ColibrÃ­"},
        "respuesta_correcta": "A"
    }
]


# Variables de seguimiento
aciertos = 0
errores = 0
preguntas_hechas = []

# Bucle principal
while aciertos < 5 and errores < 3 and len(preguntas_hechas) < len(preguntas):
    pregunta = random.choice(preguntas)

    if pregunta in preguntas_hechas:
        continue  # Saltar si ya fue hecha

    preguntas_hechas.append(pregunta)

    print(pregunta["pregunta"])
    for letra, opcion in pregunta["opciones"].items():
        print(f"  {letra}) {opcion}")

    respuesta = input("Tu respuesta (A, B, C o D): ").upper().strip()

    if respuesta == pregunta["respuesta_correcta"]:
        aciertos += 1
        print("âœ… Â¡Correcto!")
    else:
        errores += 1
        correcta = pregunta["respuesta_correcta"]
        print(f"âŒ Incorrecto. La respuesta correcta era {correcta}) {pregunta['opciones'][correcta]}")

    print(f"Aciertos: {aciertos} | Errores: {errores} | Respondidas: {len(preguntas_hechas)}\n")

# Resultado final
if aciertos == 5:
    print("ğŸ† Â¡Has ganado!")
elif errores == 3:
    print("ğŸ’€ Has perdido. IntÃ©ntalo de nuevo.")
else:
    print("ğŸ Fin de las preguntas.")
```
