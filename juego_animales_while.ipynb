# 🐾 Juego de Preguntas sobre Animales (Versión WHILE)

Este juego tipo test pone a prueba tus conocimientos sobre animales.

- Cada pregunta tiene 4 opciones (A, B, C, D)
- Solo una respuesta es correcta
- El juego termina cuando:
  - Acumulas **5 aciertos** (ganas)
  - Acumulas **3 errores** (pierdes)
  - O se acaban las preguntas disponibles


```python
import random

# Lista de preguntas
preguntas = [
    {
        "pregunta": "¿Cuál de estos animales es un mamífero?",
        "opciones": {"A": "Pingüino", "B": "Delfín", "C": "Tiburón", "D": "Cocodrilo"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "¿Qué animal tiene la lengua más larga en proporción a su cuerpo?",
        "opciones": {"A": "Jirafa", "B": "Camaleón", "C": "Hormiga", "D": "Murciélago"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "¿Cuál de estos animales pone huevos?",
        "opciones": {"A": "Gato", "B": "Perro", "C": "Gallina", "D": "Caballo"},
        "respuesta_correcta": "C"
    },
    {
        "pregunta": "¿Qué animal es conocido por su memoria?",
        "opciones": {"A": "Elefante", "B": "Cebra", "C": "Ratón", "D": "Tortuga"},
        "respuesta_correcta": "A"
    },
    {
        "pregunta": "¿Cuál es el único mamífero capaz de volar?",
        "opciones": {"A": "Pato", "B": "Murciélago", "C": "Avestruz", "D": "Gaviota"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "¿Qué animal puede cambiar de color para camuflarse?",
        "opciones": {"A": "Tiburón", "B": "Pulpo", "C": "Camaleón", "D": "Gato"},
        "respuesta_correcta": "C"
    },
    {
        "pregunta": "¿Cuál de estos animales es un reptil?",
        "opciones": {"A": "Rana", "B": "Serpiente", "C": "Pingüino", "D": "Ballena"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "¿Qué animal tiene una sola pierna cuando camina?",
        "opciones": {"A": "Canguro", "B": "Flamenco", "C": "Pulpo", "D": "Caracol"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "¿Qué animal duerme de pie?",
        "opciones": {"A": "Gato", "B": "Caballo", "C": "Loro", "D": "Oso"},
        "respuesta_correcta": "B"
    },
    {
        "pregunta": "¿Qué animal tiene el corazón más grande en proporción a su cuerpo?",
        "opciones": {"A": "Ballena azul", "B": "Elefante", "C": "Ratón", "D": "Colibrí"},
        "respuesta_correcta": "A"
    }
]


# Variables de seguimiento
aciertos = 0
errores = 0
preguntas_hechas = []

# Bucle principal
while aciertos < 5 and errores < 3 and len(preguntas_hechas) < len(preguntas):
    pregunta = random.choice(preguntas)

    if pregunta in preguntas_hechas:
        continue  # Saltar si ya fue hecha

    preguntas_hechas.append(pregunta)

    print(pregunta["pregunta"])
    for letra, opcion in pregunta["opciones"].items():
        print(f"  {letra}) {opcion}")

    respuesta = input("Tu respuesta (A, B, C o D): ").upper().strip()

    if respuesta == pregunta["respuesta_correcta"]:
        aciertos += 1
        print("✅ ¡Correcto!")
    else:
        errores += 1
        correcta = pregunta["respuesta_correcta"]
        print(f"❌ Incorrecto. La respuesta correcta era {correcta}) {pregunta['opciones'][correcta]}")

    print(f"Aciertos: {aciertos} | Errores: {errores} | Respondidas: {len(preguntas_hechas)}\n")

# Resultado final
if aciertos == 5:
    print("🏆 ¡Has ganado!")
elif errores == 3:
    print("💀 Has perdido. Inténtalo de nuevo.")
else:
    print("🏁 Fin de las preguntas.")
```
